```python
import pandas as pd
```

```python
DATASET_ABSOLUTE_PATH: str = r"E:\dataset\company\companies_sorted.csv"

dataset = pd.read_csv(DATASET_ABSOLUTE_PATH)

dataset.head()
```

```python
def create_keywords(row):
    if pd.notna(row['name']):
        keywords = row['name'].split()  
    else:
        keywords = [] 
        
    # Add industry, locality, and country as additional keywords
    # Check for NaN values and skip them
    if pd.notna(row['industry']):
        keywords.append(row['industry'])
    if pd.notna(row['locality']):
        keywords.append(row['locality'])
    if pd.notna(row['country']):
        keywords.append(row['country'])
    
    return ','.join(keywords)

dataset['keywords'] = dataset.apply(create_keywords, axis=1)
```

```python
dataset[['name', 'keywords']].head()
```

```python
dataset.to_csv('dataset.csv', columns=['name', 'keywords'], index=False)
```
